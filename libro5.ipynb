```python
import matplotlib.pyplot as plt
import numpy as np

x,y = np.loadtxt("ejemplo.csv", delimiter=',', unpack = True)
w,z = np.loadtxt("ejemplo1.csv", delimiter=',', unpack = True)

plt.plot(x,y, marker = 'o', label = 'Datos1')
plt.plot(w,z, marker = 'o', label = 'Datos2')

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-2,12)
plt.ylim(0,10)
plt.xticks(range(0,10,1))
plt.yticks(range(0,10,1))
plt.title('grafica de dos grupos de datos')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x,y = np.loadtxt("ejemplo.csv", delimiter=',', unpack = True)
w,z = np.loadtxt("ejemplo1.csv", delimiter=',', unpack = True)

plt.plot(x,y, marker = 'o', label = 'Datos1')
plt.plot(w,z,'r*-', label = 'Datos2')

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-2,12)
plt.ylim(0,10)
plt.xticks(range(0,10,1))
plt.yticks(range(0,10,1))
plt.title('grafica de dos grupos de datos')
plt.legend()
plt.show()
```

utilizzando numpy y matplotlib se pueden graficar fns matematicas definiendo en x el rango de valores donde se quiere la grafica, y en y la funcion a graficar.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-10,11.8,0.1)
y= lambda x:3*x**4+2*x**3-9*x


plt.plot(x,y(x))
plt.show()
```
